```python
class store_deaths_by_county(cv.Analyzer):

    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs) # This is necessary to initialize the class properly
        self.t = [] # List for storing the time step
        self.D_24001 = [] # List for storing deaths for each county
        self.D_24003 = []
        self.D_24005 = []
        self.D_24009 = []
        self.D_24011 = []
        self.D_24013 = []
        self.D_24015 = []
        self.D_24017 = []
        self.D_24019 = []
        self.D_24021 = []
        self.D_24023 = []
        self.D_24025 = []
        self.D_24027 = []
        self.D_24029 = []
        self.D_24031 = []
        self.D_24033 = []
        self.D_24035 = []
        self.D_24037 = []
        self.D_24039 = []
        self.D_24041 = []
        self.D_24043 = []
        self.D_24045 = []
        self.D_24047 = []
        self.D_24510 = []
        return

    def apply(self, sim):
        # Identify agents living in each county
        people_24001 = sim.people.county == 24001
        people_24003 = sim.people.county == 24003
        people_24005 = sim.people.county == 24005
        people_24009 = sim.people.county == 24009
        people_24011 = sim.people.county == 24011
        people_24013 = sim.people.county == 24013
        people_24015 = sim.people.county == 24015
        people_24017 = sim.people.county == 24017
        people_24019 = sim.people.county == 24019
        people_24021 = sim.people.county == 24021
        people_24023 = sim.people.county == 24023
        people_24025 = sim.people.county == 24025
        people_24027 = sim.people.county == 24027
        people_24029 = sim.people.county == 24029
        people_24031 = sim.people.county == 24031
        people_24033 = sim.people.county == 24033
        people_24035 = sim.people.county == 24035
        people_24037 = sim.people.county == 24037
        people_24039 = sim.people.county == 24039
        people_24041 = sim.people.county == 24041
        people_24043 = sim.people.county == 24043
        people_24045 = sim.people.county == 24045
        people_24047 = sim.people.county == 24047
        people_24510 = sim.people.county == 24510
        # Identify agents who have died
        people_dead = sim.people.dead
        # Identify agents from each county who have died
        people_dead_24001 = cv.true(people_24001 * sim.people.dead)
        people_dead_24003 = cv.true(people_24003 * sim.people.dead)
        people_dead_24005 = cv.true(people_24005 * sim.people.dead)
        people_dead_24009 = cv.true(people_24009 * sim.people.dead)
        people_dead_24011 = cv.true(people_24011 * sim.people.dead)
        people_dead_24013 = cv.true(people_24013 * sim.people.dead)
        people_dead_24015 = cv.true(people_24015 * sim.people.dead)
        people_dead_24017 = cv.true(people_24017 * sim.people.dead)
        people_dead_24019 = cv.true(people_24019 * sim.people.dead)
        people_dead_24021 = cv.true(people_24021 * sim.people.dead)
        people_dead_24023 = cv.true(people_24023 * sim.people.dead)
        people_dead_24025 = cv.true(people_24025 * sim.people.dead)
        people_dead_24027 = cv.true(people_24027 * sim.people.dead)
        people_dead_24029 = cv.true(people_24029 * sim.people.dead)
        people_dead_24031 = cv.true(people_24031 * sim.people.dead)
        people_dead_24033 = cv.true(people_24033 * sim.people.dead)
        people_dead_24035 = cv.true(people_24035 * sim.people.dead)
        people_dead_24037 = cv.true(people_24037 * sim.people.dead)
        people_dead_24039 = cv.true(people_24039 * sim.people.dead)
        people_dead_24041 = cv.true(people_24041 * sim.people.dead)
        people_dead_24043 = cv.true(people_24043 * sim.people.dead)
        people_dead_24045 = cv.true(people_24045 * sim.people.dead)
        people_dead_24047 = cv.true(people_24047 * sim.people.dead)
        people_dead_24510 = cv.true(people_24510 * sim.people.dead)
        # At each time step, count agents who have died for each county and append to the corresponding list
        self.t.append(sim.t)
        self.D_24001.append(len(people_dead_24001))
        self.D_24003.append(len(people_dead_24003))
        self.D_24005.append(len(people_dead_24005))
        self.D_24009.append(len(people_dead_24009))
        self.D_24011.append(len(people_dead_24011))
        self.D_24013.append(len(people_dead_24013))
        self.D_24015.append(len(people_dead_24015))
        self.D_24017.append(len(people_dead_24017))
        self.D_24019.append(len(people_dead_24019))
        self.D_24021.append(len(people_dead_24021))
        self.D_24023.append(len(people_dead_24023))
        self.D_24025.append(len(people_dead_24025))
        self.D_24027.append(len(people_dead_24027))
        self.D_24029.append(len(people_dead_24029))
        self.D_24031.append(len(people_dead_24031))
        self.D_24033.append(len(people_dead_24033))
        self.D_24035.append(len(people_dead_24035))
        self.D_24037.append(len(people_dead_24037))
        self.D_24039.append(len(people_dead_24039))
        self.D_24041.append(len(people_dead_24041))
        self.D_24043.append(len(people_dead_24043))
        self.D_24045.append(len(people_dead_24045))
        self.D_24047.append(len(people_dead_24047))
        self.D_24510.append(len(people_dead_24510))
        return

    def df(self):
        # Identify all agents (not just deceased) by county so we can use the counts to calculate per 100k values
        n_people_24001 = len(people.loc[people['county'] == 24001])
        n_people_24003 = len(people.loc[people['county'] == 24003])
        n_people_24005 = len(people.loc[people['county'] == 24005])
        n_people_24009 = len(people.loc[people['county'] == 24009])
        n_people_24011 = len(people.loc[people['county'] == 24011])
        n_people_24013 = len(people.loc[people['county'] == 24013])
        n_people_24015 = len(people.loc[people['county'] == 24015])
        n_people_24017 = len(people.loc[people['county'] == 24017])
        n_people_24019 = len(people.loc[people['county'] == 24019])
        n_people_24021 = len(people.loc[people['county'] == 24021])
        n_people_24023 = len(people.loc[people['county'] == 24023])
        n_people_24025 = len(people.loc[people['county'] == 24025])
        n_people_24027 = len(people.loc[people['county'] == 24027])
        n_people_24029 = len(people.loc[people['county'] == 24029])
        n_people_24031 = len(people.loc[people['county'] == 24031])
        n_people_24033 = len(people.loc[people['county'] == 24033])
        n_people_24035 = len(people.loc[people['county'] == 24035])
        n_people_24037 = len(people.loc[people['county'] == 24037])
        n_people_24039 = len(people.loc[people['county'] == 24039])
        n_people_24041 = len(people.loc[people['county'] == 24041])
        n_people_24043 = len(people.loc[people['county'] == 24043])
        n_people_24045 = len(people.loc[people['county'] == 24045])
        n_people_24047 = len(people.loc[people['county'] == 24047])
        n_people_24510 = len(people.loc[people['county'] == 24510])
        # Generate a dictionary of cumulative counts by subgroup
        d = {'day':self.t, 
            'D_24001_cum':self.D_24001,
            'D_24003_cum':self.D_24003,
            'D_24005_cum':self.D_24005,
            'D_24009_cum':self.D_24009,
            'D_24011_cum':self.D_24011,
            'D_24013_cum':self.D_24013,
            'D_24015_cum':self.D_24015,
            'D_24017_cum':self.D_24017,
            'D_24019_cum':self.D_24019,
            'D_24021_cum':self.D_24021,
            'D_24023_cum':self.D_24023,
            'D_24025_cum':self.D_24025,
            'D_24027_cum':self.D_24027,
            'D_24029_cum':self.D_24029,
            'D_24031_cum':self.D_24031,
            'D_24033_cum':self.D_24033,
            'D_24035_cum':self.D_24035,
            'D_24037_cum':self.D_24037,
            'D_24039_cum':self.D_24039,
            'D_24041_cum':self.D_24041,
            'D_24043_cum':self.D_24043,
            'D_24045_cum':self.D_24045,
            'D_24047_cum':self.D_24047,
            'D_24510_cum':self.D_24510,}
        # Make a data frame from that dictionary
        df = pd.DataFrame(data=d)
        # Generate daily values by taking the cumulative difference between one day and the previous
        df['D_24001_new'] = df['D_24001_cum'].diff()
        df['D_24003_new'] = df['D_24003_cum'].diff()
        df['D_24005_new'] = df['D_24005_cum'].diff()
        df['D_24009_new'] = df['D_24009_cum'].diff()
        df['D_24011_new'] = df['D_24011_cum'].diff()
        df['D_24013_new'] = df['D_24013_cum'].diff()
        df['D_24015_new'] = df['D_24015_cum'].diff()
        df['D_24017_new'] = df['D_24017_cum'].diff()
        df['D_24019_new'] = df['D_24019_cum'].diff()
        df['D_24021_new'] = df['D_24021_cum'].diff()
        df['D_24023_new'] = df['D_24023_cum'].diff()
        df['D_24025_new'] = df['D_24025_cum'].diff()
        df['D_24027_new'] = df['D_24027_cum'].diff()
        df['D_24029_new'] = df['D_24029_cum'].diff()
        df['D_24031_new'] = df['D_24031_cum'].diff()
        df['D_24033_new'] = df['D_24033_cum'].diff()
        df['D_24035_new'] = df['D_24035_cum'].diff()
        df['D_24037_new'] = df['D_24037_cum'].diff()
        df['D_24039_new'] = df['D_24039_cum'].diff()
        df['D_24041_new'] = df['D_24041_cum'].diff()
        df['D_24043_new'] = df['D_24043_cum'].diff()
        df['D_24045_new'] = df['D_24045_cum'].diff()
        df['D_24047_new'] = df['D_24047_cum'].diff()
        df['D_24510_new'] = df['D_24510_cum'].diff()
        # Calculate cumulative per 100k values
        df['D_24001_cum_per100k'] = df['D_24001_cum'] / n_people_24001 * 100_000
        df['D_24003_cum_per100k'] = df['D_24003_cum'] / n_people_24003 * 100_000
        df['D_24005_cum_per100k'] = df['D_24005_cum'] / n_people_24005 * 100_000
        df['D_24009_cum_per100k'] = df['D_24009_cum'] / n_people_24009 * 100_000
        df['D_24011_cum_per100k'] = df['D_24011_cum'] / n_people_24011 * 100_000
        df['D_24013_cum_per100k'] = df['D_24013_cum'] / n_people_24013 * 100_000
        df['D_24015_cum_per100k'] = df['D_24015_cum'] / n_people_24015 * 100_000
        df['D_24017_cum_per100k'] = df['D_24017_cum'] / n_people_24017 * 100_000
        df['D_24019_cum_per100k'] = df['D_24019_cum'] / n_people_24019 * 100_000
        df['D_24021_cum_per100k'] = df['D_24021_cum'] / n_people_24021 * 100_000
        df['D_24023_cum_per100k'] = df['D_24023_cum'] / n_people_24023 * 100_000
        df['D_24025_cum_per100k'] = df['D_24025_cum'] / n_people_24025 * 100_000
        df['D_24027_cum_per100k'] = df['D_24027_cum'] / n_people_24027 * 100_000
        df['D_24029_cum_per100k'] = df['D_24029_cum'] / n_people_24029 * 100_000
        df['D_24031_cum_per100k'] = df['D_24031_cum'] / n_people_24031 * 100_000
        df['D_24033_cum_per100k'] = df['D_24033_cum'] / n_people_24033 * 100_000
        df['D_24035_cum_per100k'] = df['D_24035_cum'] / n_people_24035 * 100_000
        df['D_24037_cum_per100k'] = df['D_24037_cum'] / n_people_24037 * 100_000
        df['D_24039_cum_per100k'] = df['D_24039_cum'] / n_people_24039 * 100_000
        df['D_24041_cum_per100k'] = df['D_24041_cum'] / n_people_24041 * 100_000
        df['D_24043_cum_per100k'] = df['D_24043_cum'] / n_people_24043 * 100_000
        df['D_24045_cum_per100k'] = df['D_24045_cum'] / n_people_24045 * 100_000
        df['D_24047_cum_per100k'] = df['D_24047_cum'] / n_people_24047 * 100_000
        df['D_24510_cum_per100k'] = df['D_24510_cum'] / n_people_24510 * 100_000
        # Calculate daily per 100k values
        df['D_24001_new_per100k'] = df['D_24001_new'] / n_people_24001 * 100_000
        df['D_24003_new_per100k'] = df['D_24003_new'] / n_people_24003 * 100_000
        df['D_24005_new_per100k'] = df['D_24005_new'] / n_people_24005 * 100_000
        df['D_24009_new_per100k'] = df['D_24009_new'] / n_people_24009 * 100_000
        df['D_24011_new_per100k'] = df['D_24011_new'] / n_people_24011 * 100_000
        df['D_24013_new_per100k'] = df['D_24013_new'] / n_people_24013 * 100_000
        df['D_24015_new_per100k'] = df['D_24015_new'] / n_people_24015 * 100_000
        df['D_24017_new_per100k'] = df['D_24017_new'] / n_people_24017 * 100_000
        df['D_24019_new_per100k'] = df['D_24019_new'] / n_people_24019 * 100_000
        df['D_24021_new_per100k'] = df['D_24021_new'] / n_people_24021 * 100_000
        df['D_24023_new_per100k'] = df['D_24023_new'] / n_people_24023 * 100_000
        df['D_24025_new_per100k'] = df['D_24025_new'] / n_people_24025 * 100_000
        df['D_24027_new_per100k'] = df['D_24027_new'] / n_people_24027 * 100_000
        df['D_24029_new_per100k'] = df['D_24029_new'] / n_people_24029 * 100_000
        df['D_24031_new_per100k'] = df['D_24031_new'] / n_people_24031 * 100_000
        df['D_24033_new_per100k'] = df['D_24033_new'] / n_people_24033 * 100_000
        df['D_24035_new_per100k'] = df['D_24035_new'] / n_people_24035 * 100_000
        df['D_24037_new_per100k'] = df['D_24037_new'] / n_people_24037 * 100_000
        df['D_24039_new_per100k'] = df['D_24039_new'] / n_people_24039 * 100_000
        df['D_24041_new_per100k'] = df['D_24041_new'] / n_people_24041 * 100_000
        df['D_24043_new_per100k'] = df['D_24043_new'] / n_people_24043 * 100_000
        df['D_24045_new_per100k'] = df['D_24045_new'] / n_people_24045 * 100_000
        df['D_24047_new_per100k'] = df['D_24047_new'] / n_people_24047 * 100_000
        df['D_24510_new_per100k'] = df['D_24510_new'] / n_people_24510 * 100_000
        return df
```

## 6.2 Run the sim with the analyzer
Using the parameters from `5_calibration.ipynb`, run the sim and call the analyzer. Save the sim and resulting analyzer dataframe so you can load them later. The analyzer dataframe includes cumulative and daily deaths per 100k for each county.

```python
people = pd.read_csv(f'{path}/people_all.csv',low_memory=False) # Read in people_all.csv so we have its length
```

```python
pars = dict( # Use the same parameters you decided on in 5_calibration
    pop_size = len(people), 
    pop_infected = 300, 
    start_day = '2020-02-01', 
    end_day   = '2020-06-30',
    rel_death_prob = 0.1, 
    beta = 0.032,) 

sim = cv.Sim(pars, popfile=f'{path}/greasypop.ppl', analyzers=store_deaths_by_county(label='deaths_by_county')) # Initialize sim with analyzer
sim.run() # Run sim
sim.save(f'{path}/greasypop_by_county.sim') # Save sim
store_deaths_by_county = sim.get_analyzer('deaths_by_county') # Retrieve analyzer by label
sim_deaths = store_deaths_by_county.df() # Get the dataframe from the analyzer
sim_deaths.to_csv(f'{path}/greasypop_by_county.csv') # Save the dataframe
sim_deaths # Preview data
```

## 6.5 Map deaths by county for day 100
In `4_calibration.ipynb`, we saw that peak deaths for our entire synthetic population occured around May 10, which is day 100 in the sim. Now we will map cumulative simulated and observed deaths per 100k for each county on day 100. First, reformat the data and compare the values. 

```python
sim_deaths = pd.read_csv(f'{path}/greasypop_by_county.csv')
sim_deaths = sim_deaths.drop(columns=['Unnamed: 0','day']) # Drop columns not needed
sim_deaths = sim_deaths.T # Transpose the table so day numbers are columns
sim_deaths = sim_deaths.iloc[48:72,:][[100]].reset_index() # Keep only cumulative deaths per 100k for each county on day 100
sim_deaths = sim_deaths.rename(columns={'index':'county_fip',100:'sim_cum_per100k'}) # Rename columns
sim_deaths['county_fip'] = sim_deaths['county_fip'].str[4:7].astype(int).astype(str) # Reformat so county_fip will merge with shapefile data

obs_deaths = pd.read_csv(f'{path}/observed_by_county.csv') # Read in observed deaths by county
obs_deaths = obs_deaths.drop(columns=['Unnamed: 0','level_0','index']) # Drop columns not needed
obs_deaths = obs_deaths.T # Transpose the table so day numbers are columns
obs_deaths = obs_deaths.iloc[:24,:][[100]].reset_index() # Keep only cumulative deaths per 100k for each county on day 100
obs_deaths = obs_deaths.rename(columns={'index':'county_fip',100:'obs_cum_per100k'}) # Rename columns
obs_deaths['county_fip'] = obs_deaths['county_fip'].str[2:5].astype(int).astype(str) # Reformat so county_fip will merge with shapefile data

deaths = sim_deaths.merge(obs_deaths,how='left',on='county_fip') # merge sim and obs deaths
deaths # Preview data
```

### 6.5.1 Map simulated deaths by county

```python
county_shape = gpd.read_file(f'{path}/shape_county/BNDY_CountyPhyBoundaryGen_DoIT.shp')
county_shape['county_fip'] = county_shape['county_fip'].astype(str)

map_sim_deaths = county_shape.merge(deaths, on='county_fip', how='left')
map_sim_deaths['sim_cum_per100k'] = map_sim_deaths['sim_cum_per100k'].astype(float)

fig, ax = plt.subplots(1, 1, figsize=(16, 6))
map_sim_deaths.plot(column='sim_cum_per100k', cmap='summer_r', legend=True, ax=ax)
ax.set_title('Simulated cumulative deaths per 100k by county\n2020-05-10')
ax.axis('off')
for idx, row in map_sim_deaths.iterrows(): # Add values in the center of shapes
    plt.annotate(text=round(row['sim_cum_per100k']), xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                 horizontalalignment='center', fontsize=8, color='black')
plt.show()
```

### 6.5.2 Map observed deaths by county

```python
county_shape = gpd.read_file(f'{path}/shape_county/BNDY_CountyPhyBoundaryGen_DoIT.shp')
county_shape['county_fip'] = county_shape['county_fip'].astype(str)

map_obs_deaths = county_shape.merge(obs_deaths, on='county_fip', how='left')
map_obs_deaths['obs_cum_per100k'] = map_obs_deaths['obs_cum_per100k'].astype(float)

fig, ax = plt.subplots(1, 1, figsize=(16, 6))
map_obs_deaths.plot(column='obs_cum_per100k', cmap='summer_r', legend=True, ax=ax)
ax.set_title('Observed cumulative deaths per 100k by county\n2020-05-10')
ax.axis('off')
for idx, row in map_obs_deaths.iterrows(): # Add values in the center of shapes
    plt.annotate(text=round(row['obs_cum_per100k']), xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                 horizontalalignment='center', fontsize=8, color='black')
plt.show()
```
